```python
res = []

for r1, r2 in zip(t1.itertuples(), t2.itertuples()):
    if r1.end - r1.start < r2.end - r2.start:
        res.append((r1.id, r1.start, r1.end))
    else:
        res.append((r2.id, r2.start, r2.end))
```

```python
t5 = pd.DataFrame(data=res, columns=["id", "start", "end"])
```

```python
t5.to_csv("../predictions/t2_roberta_v1+v2.csv", index=False)
```
