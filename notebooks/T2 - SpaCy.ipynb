```python
import json
import spacy
import pandas as pd
from spacy.tokens import DocBin
```

```python
with open("../data/sentiment/train.json") as infile:
    data = json.load(infile)

df = pd.DataFrame(data=data)

with open("../data/sentiment/testing.json") as infile:
    data = json.load(infile)

test = pd.DataFrame(data=data)
```

```python
train = df[~df.text.isin(test.text)].drop_duplicates("text")
valid = df[df.text.isin(test.text)]
```

```python
print(len(train), len(valid))
```

```python
df
```

```python
nlp = spacy.blank("en")
```

```python
def r(s):
    return "negative" if s == "positive" else "positive"

# the DocBin will store the example documents
train_db = DocBin()
for row in train.itertuples():
    doc = nlp(row.text)
    start = row.start_char_pos
    
    if row.end_char_pos < 1:
        end = len(row.text) - row.end_char_pos
    elif row.end_char_pos >= len(row.text):
        end = len(row.text)
    else:
        end = row.end_char_pos

    span = doc.char_span(start, end, label=row.sentiment)
    if not span:
        print(row.text)
        continue
    doc.ents = [span]
    doc.spans["sc"] = list(doc.ents)
    train_db.add(doc)

train_db.to_disk("../spacy/train.spacy")
```

```python
# the DocBin will store the example documents
valid_db = DocBin()
for row in valid.itertuples():
    doc = nlp(row.text)
    start = row.start_char_pos
    end = row.end_char_pos if row.end_char_pos > 0 else len(row.text) - row.end_char_pos
    span = doc.char_span(start, end, label=row.sentiment)
    if not span:
        print(row.text)
        continue
    doc.ents = [span]
    doc.spans["sc"] = list(doc.ents)
    valid_db.add(doc)

valid_db.to_disk("../spacy/valid.spacy")
```

```python
print(len(train_db), len(valid_db))
```
